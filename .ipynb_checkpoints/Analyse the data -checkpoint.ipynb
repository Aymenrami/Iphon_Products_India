```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"C:\Users\Lenovo\Desktop\Project Aymen\Project Anlayse des Ventes IPHON\Product_Iphon_Clean.xlsx")
df.head()
```

```python
#the best 10 Phone in Rating
highest_rated =df.sort_values(by=['Sta_ Rating'],ascending=False)
highest_rated = highest_rated.head(10)
highest_rated['Product_Name']
```

```python
Iphone = highest_rated['Product_Name']
Label=Iphone.index
counts=highest_rated['Number_Of_Ratings']
```

```python
sns.barplot(x=Label,y= counts,data = highest_rated , palette = 'flare')
plt.xticks(rotation=90)
plt.show()
```

```python
MyGraph = sns.lmplot(x='Number_Of_Ratings',y='Sale_Price' ,data=df )
plt.show()
```
